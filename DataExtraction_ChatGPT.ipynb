
## Transform the output to tabular format

```python
df = pd.read_csv('ChatGPT_Output.csv')
df = df[df['Role'] =='assistant']
```

```python
texts = []

for i,n in enumerate(df['Content']):
    texts.append(n)
```

```python
columns = ['Quantity','Artefact', 'Brand', 'Model Number', 'Artefact Problem', 'Problem Symptoms', 'Problem Investigation', 'Artefacts Needed for Problem Investigation', 'Problem Resolution','Way of Carrying Out the Problem Resolution','Artefacts Needed for Problem Resolution']

output_df = pd.DataFrame(columns=columns)
```

```python
for i,text in enumerate(texts):
    print(f">>>>>>>>>>> Entry {i}")
    error = ""
    
    #i = i.strip()
    
    system_context = """You are a home maintenance expert.

You organize and provide precise technical information. 

You write as few words as possible.

If you do not have the information, you put N/A into a table."""

    
    user_request = f""" 
    text:
    
    ---
    {text}
    ---
    
This table is meant to organize information about appliances’ issues, how to investigate them, and how to solve them. You must populate the table according to the following guidelines. 



If you don’t have the information that is relevant for filling a given cell, leave that cell empty.

The kind of appliance under consideration must be put under the column with header “Artefact”. A kind of appliance can be denoted either by a simple noun - like “refrigerator” or “oven” - or by a compound phrase - like “microwave oven” or “bottom mount refrigerator”. In any case, the column with header “Artefact” must contain the entire designation, but without the brand and the serial number.

The brand of the appliance under consideration must be put under the column with header “Brand”.

The model number of the appliance under consideration must be put under the column with header “Model Number”.

The information about the issues under consideration must be put in the columns with the headers “Artefact Problem” and “Problem Symptoms”. If there is only one simple issue or faulty condition being described, that must be put under “Artefact Problem”, and the cell under “Problem Symptoms” must be left empty. If the information about the issue is more articulated, the part about the root or the underlying problem must be put under “Artefact Problem”, whereas the part about the manifestations or observable effects of the issue must be put under “Problem Symptoms”. 

Generally, a symptom can be defined as an observable or manifest effect of an underlying problem. If the information about the issue is articulated, then the symptoms must be put under “Problem Symptoms” and the underlying problem or issue under “Artefact Problem”. By contrast, if the information about the issue is very basic, it must be put entirely under “Artefact Problem”, leaving the cell under “Problem Symptoms” empty.

For example, if the text talks about refrigerators that leak fluid when their compressor pump is broken, the column with header “Artefact Problem” must be filled with “broken compressor pump”, whereas the column with header “Problem Symptoms” must be filled with “leaking fluid”.

By contrast, if the text is only talking about refrigerators that leak fluid, without any other information about the issue, then the column with header “Artefact Problem” must be filled with “leaking fluid”, whereas the column with header “Problem Symptoms” must be left empty.

If the text contains tips or instructions about the actions to perform to investigate the issue, those instructions must be put under “Problem Investigation”.

If the text also mentions tools or devices that are needed to run the investigation, those must be reported separately in the column with header “Artefacts Needed for Problem Investigation”.

If the text contains tips or instructions on how to fix or solve the issue, those instructions must be put in the columns with the headers “Problem Resolution” and “Way of Carrying Out the Problem Resolution”. If the relevant information only consists of a description of the kind of action to perform to fix or solve the issue, that information must be put in the column with the header “Problem Resolution”. If the information also specifies some method to perform that action, that method must be reported under “Way of Carrying out Problem Resolution”. 

The information that is put under “Problem Resolution” must consist of a kind of action - typically denoted by a verb or verbal phrase - with some specified object or target of that action. Examples of problem resolution descriptions include “replace the water valve” and “clean the compressor”.

A method consists of a series of simpler tasks by means of which one can complete the action reported under “Problem Resolution”. If they are mentioned at all in the text, those simpler tasks must be reported under “Way of Carrying out Problem Resolution”. 

For example, consider a text that first says that to solve a certain refrigerator issue one needs to replace the broken compressor, and then gives step-by-step instructions on how to replace the broken compressor. In this case, “replace the broken compressor” must be put under “Problem Resolution”, and the step-by-step instructions on how to replace the broken compressor that must be put under “Way of Carrying out Problem Resolution”.

If the text also mentions tools or devices that are needed to carry out the resolution, those must be reported separately in the column with header “Artefacts Needed for Problem Resolution”.

The column with header “Quantity” must be filled depending on the level of generality of the information that is being put on each row. If what is being said is true of all the appliances of a given kind, the cell under “Quantity” must be filled with “Universally”. If what is being said is true only for the most or typically of the appliances of a given kind, the cell under “Quantity” must be filled with “Typically”. If what is being said can be true or is possibly true about the appliances of a given kind, the cell under “Quantity” must be filled with “Possibly”. If the subject of what is being said is an individual appliance - like say, Emma’s refrigerator or Joe’s oven - then the cell under “Quantity” must be filled with “Particular”.



    Extract key information from that text into the following table:
|{"|".join(columns)}|
|{"---|"*len(columns)}
"""

    #print(text)
    response = openai.Completion.create(
        engine = "text-davinci-003",
        prompt = system_context + user_request,
        temperature = 0.6,
        max_tokens = 500
    )

    response_text = response.choices[0].text
    #print('response_text')
    print(response_text)

    rows = response_text.split('\n')

    data = []
    for row in rows:
        if row.startswith("|"):
            row_as_list = row.strip('|').split('|')
            if len(row_as_list) == len(columns):
                data.append(row_as_list)
            else:
                error += f'>>> Detected {len(row_as_list)} instead of {len(columns)} columns in:\n{row}'
        else:
            error += f'>>> Badly formatted row in response from entry {i}:\n{row}'

    data_df = pd.DataFrame(data, columns=columns)

    output_df = pd.concat([output_df, data_df], axis=0, ignore_index=True)

    if len(error)> 0:
        print(f">>> Original Entry Text\n{i}\n" + f">>> GPT Response Text\n{response_text}\n" + error)
```
